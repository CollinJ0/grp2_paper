```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('../data_processing/data/persistent-clonotypes/D103-df.csv', index_col=0)
df2=pd.read_csv('../data_processing/data/persistent-clonotypes/327059-df.csv', index_col=0)
```

```python
df3= pd.read_csv('../data_processing/data/persistent-sequences/D103-df.csv')
df4= pd.read_csv('../data_processing/data/persistent-sequences/327059-df.csv')
```

```python
# initialize plot
sns.set_style('white')
plt.figure(figsize=(5, 5))

# plot rarefaction data
plt.plot([0] + df1['size'], [0] + df1['average'], c='#4B60E9', label='D103', linewidth=2, alpha=0.9)
plt.plot([0] + df2['size'], [0] + df2['average'], c='#FFA862', label='327059', linewidth=2, alpha=0.9)
plt.plot([0] + df3['size'], [0] + df3['average'], c='#4B60E9', linewidth=2, alpha=0.9, linestyle=':')
plt.plot([0] + df4['size'], [0] + df4['average'], c='#FFA862', linewidth=2, alpha=0.9, linestyle=':')

# plot diagonal reference line
# plt.plot((0, 1), (1000000, 1000000), 'k:', linewidth=1)

# style the plot
ax = plt.gca()
# plot legend
ax.legend(loc='upper left', fontsize=11.5, frameon=False)
# set axis limits and labels
# ax.set_xlim((0, 1.05))
ax.set_ylim((0, 1.05))
ax.set_xlabel('sampling depth', fontsize=14)
ax.set_ylabel('% shared across timepoints', fontsize=14)
# style ticks
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', which='major', labelsize=12, length=6, width=1.25, pad=12, right=False)
# hide top, left and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# save or show
# plt.show()
plt.tight_layout()
plt.savefig('../make_figures/figures/percent_shared_overtime.pdf', dpi=300)
```
